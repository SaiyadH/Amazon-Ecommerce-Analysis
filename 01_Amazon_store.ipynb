#### Data abhi clean nhi hai datatype me issues hai

```python
def clean_data(st):

    text_to_number = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
    cleaned_data = []
    unique_user = set()
    for user in st:
        # clean rating
        raw_rating = str(user['rating']).strip().lower()

        if (raw_rating in text_to_number):
            raw_rating = text_to_number[raw_rating]

        user['rating'] = raw_rating

        # Handle are missing values
        raw_age = user.get("age")
        if raw_age == None:
            user["age"] = None

        # deduplication
        if (user['name'].strip() in unique_user):
            continue

        unique_user.add(user['name'])
        cleaned_data.append(user)
    return cleaned_data


        
      
```

```python
data = clean_data(st)
```

```python
def get_insights(st):

    # avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user['rating'])

    print(f"Avarage rating {tot_rating/len(data)}")

    # pecentage of user with poor rating
    poor_rating = 0
    for user in data:
        if (float(user['rating']) < 3):
            poor_rating += 1

    print(f"Percentage or Poor rating :{poor_rating/len(data) * 100}%")
```

```python
get_insights(data)
```

```python
# Build recommendations systems
def get_recommendation(data):
    recommendation = []

    for user in data:
        curr_recom = {}
        curr_recom['name'] = user['name']

        if (float(user['rating']) >= 4 ):
            curr_recom['brand'] = 'Apple'
        else:
            curr_recom['brand'] = "samsung"

        recommendation.append(curr_recom)

    return recommendation
```
